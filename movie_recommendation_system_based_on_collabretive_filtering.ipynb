```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
count_matrix = vectorizer.fit_transform(dataset["combined_feature"])
```

```python
print(vectorizer.get_feature_names())
```

```python
print(count_matrix.toarray())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
simlarity_score = cosine_similarity(count_matrix)
print(simlarity_score)
```

```python
movie_user_likes = "Avatar"
```

```python
# Get Index of the movie from the dataset
def get_movie_index(movie_user_likes):
    for i in range(len(dataset['movie_title'])):
        if dataset['movie_title'][i] == movie_user_likes.lower():
            return i
```

```python
movie_index = get_movie_index(movie_user_likes)
```

```python
s_score = list(enumerate(simlarity_score[movie_index]))
```

```python
s_score
```

```python
s_score_sorted = sorted(s_score, key = lambda x: x[1], reverse = True)
```

```python
s_score_sorted
```

```python
# Top 10 similar movies
indexes = [t[0] for t in s_score_sorted[0:10]]
```

```python
indexes
```

```python
# Get Movie Title from Index
def get_movie_names(indexes):
    movie_match = []
    for index in indexes:
        movie_match.append(dataset['title'][index].title())
    return movie_match
```

```python
top_10_movie_match = get_movie_names(indexes)
```
