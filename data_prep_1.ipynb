```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("movie_metadata.csv")
```

```python
dataset.title_year.value_counts(dropna=False).sort_index().plot(kind='barh',figsize=(20,30))
plt.show()
```

```python
# recomendation will be based on these features
data = dataset.loc[:, ['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
```

```python
data.head()
```

```python
data.info()
```

```python
data['actor_1_name'] = data['actor_1_name'].replace(np.nan, 'unknown')
data['actor_2_name'] = data['actor_2_name'].replace(np.nan, 'unknown')
data['actor_3_name'] = data['actor_3_name'].replace(np.nan, 'unknown') 
data['director_name'] = data['director_name'].replace(np.nan, 'unknown')
```

```python
data
```

```python
data["genres"] = data["genres"].replace("|", " ")
```

```python
data
```

```python
data
```

```python
data['movie_title'] = data['movie_title'].str.lower()
```

```python
data["movie_title"][1]
```

```python
# removing end character
data['movie_title'] = data['movie_title'].apply(lambda x: x[:-1])
```

```python
data["movie_title"][1]
```

```python
data.to_csv('data.csv',index=False)
```
